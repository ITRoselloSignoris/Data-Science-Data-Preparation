## Consigna 1

### 0. Importar bibliotecas

```python
import pandas as pd
from funpymodeling import status
```

### 1. Carga de datos

```python
data = pd.read_csv("Datasets/encuesta-anual-hogares-2019.csv",sep=",")
data
```

### 1.Inspeccion Inicial

```python
data.head(5)
```

```python
status(data)
```

```python
status(data).query("q_nan > 0")
```

```python
data2 = data.drop(columns=["id", "hijos_nacidos_vivos"])
data2
```

### 3.Discretizacion

```python
data2["ingresos_familiares_cat"] = pd.qcut(data2["ingresos_familiares"],q=8)
data2["ingresos_familiares_cat"]
```

```python
data2["ingreso_per_capita_familiar_cat"] = pd.qcut(data2["ingreso_per_capita_familiar"],q=10)
data2["ingreso_per_capita_familiar_cat"]
```

```python
data2["ingreso_total_lab"] = pd.qcut(data2["ingreso_total_lab"], q=10, duplicates="drop")
data2["ingreso_total_no_lab"] = pd.qcut(data2["ingreso_total_no_lab"], q=4, duplicates="drop")
```

```python
data2["edad_cat"] = pd.cut(data2["edad"],bins=5)
data2["edad_cat"]
```

```python
status(data2)
```

### 4.Preparacion de datos

```python
data2["comuna"] = data2["comuna"].astype(str)
data2["nhogar"] = data2["nhogar"].astype(str)
```

```python
status(data2["comuna"])
```

```python
data2["años_escolaridad"].unique()
```

```python
data2["años_escolaridad"] = data2["años_escolaridad"].replace("Ningun año de escolaridad aprobado", "0")
data2["años_escolaridad"].unique()
```

```python
data2["años_escolaridad"] = data2["años_escolaridad"].astype(float).astype("Int32")
status(data2["años_escolaridad"])
```

```python
data2["años_escolaridad_cat"] = pd.qcut(data2["años_escolaridad"], q=5)
data2["años_escolaridad_cat"]
```

```python
data2 = data2.dropna(subset=["situacion_conyugal","sector_educativo","lugar_nacimiento","afiliacion_salud"])
status(data2).query("q_nan > 0")
```

```python
data2 = data2.reset_index(drop=True)
```

```python
data2["años_escolaridad_cat"] = data2["años_escolaridad_cat"].cat.add_categories("desconocido")
data2["años_escolaridad_cat"]
```

```python
data2["años_escolaridad_cat"] = data2["años_escolaridad_cat"].fillna(value="desconocido")
status(data2["años_escolaridad_cat"])
```

```python
status(data2)
```

```python
data2.drop(columns=["años_escolaridad","edad","ingreso_per_capita_familiar", "ingresos_familiares"],axis=1,inplace=True)
```

```python
data2["nivel_max_educativo"] = data2["nivel_max_educativo"].fillna(value="desconocido")
status(data2["nivel_max_educativo"])
```

```python
status(data2)
```

### 5. One Hot Encoding

```python
data_ohe = pd.get_dummies(data2)
data_ohe
```
